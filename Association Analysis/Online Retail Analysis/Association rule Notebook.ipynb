# preprocessing

mengubah data kebentuk one hot dengan tiap baris / row merepresentasikan satu transaksi / invoice dari barang barang yang dibeli

```python
dataset = UK.groupby(['Invoice'])['Description'].value_counts()
dataset_df = dataset.unstack(fill_value=0)
dataset_df.reset_index(inplace=True)
```

```python
dataset_df.head(20)
```

```python
cleanData = dataset_df.set_index('Invoice')
```

```python
cleanData.head()
```

```python
# memastika Encoding data antara 1 atau 0
def encode_units(x):
    return 1 if x >= 1 else 0

cleanData= cleanData.applymap(encode_units)
```

# Menggunakan mlxtend untuk membuat melakukan Association Analysis dengan Apriori

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth, association_rules
```

# 1. Apriori
1. Candidate Generation
2. Brute Force - (mencoba semua kombinasi yang mungkin)
3. Multiple scanning

# 1. Mining frequent Item set
Frequent itemset adalah itemset yang muncul dalam dataset dengan frekuensi yang lebih tinggi dari ambang batas yang ditentukan.

semua kombonsai itemset dengan minimal support yaitu 0.02

```python
freq_item = apriori(cleanData, min_support=0.02, use_colnames=True)
freq_item
```

# 2. Rules Generation

**Confidence** mengukur persentase keranjang(basket) yang mengandung produk di sisi kiri juga mengandung produk di sisi kanan.

- Seberapa sering item X dan Y muncul bersama berdasarkan jumlah X terjadi (item kiri)
- confidence membantu untuk melihat seberapa valid sebuah aturan


Association rule dengan minimal support yaitu 0.02

```python
hasil = association_rules(freq_item, metric='confidence', min_threshold= 0.02)
hasil
```

```python
hasil.to_csv('rules.csv')
```

# Make an association rule mining out of this dataset

- Input: Stock code / Item Description
- Output: Max 3 items recommended out of the input

```python
rules = pd.read_csv("rules.csv")   # hasil dari Association rule

def recommend_items(input_items, rules):
    recommendations = []
    
   
    # Filter rules for the input item
    filtered_rules = rules[rules['antecedents'].str.contains(input_items)]
    
    # Sort by confidence and lift
    sorted_rules = filtered_rules.sort_values(by=['confidence', 'lift'], ascending=[False, False])
    
    # Extract consequents
    consequents = sorted_rules['consequents'].tolist()
    
    # Remove duplicates and limit recommendations
    return consequents
```

barang lain yang sering dibeli bersamaan ketika sebuah barang dibeli

```python
# Input example 
# 60 TEATIME FAIRY CAKE CASES
input_items = "60 TEATIME FAIRY CAKE CASES"
output = recommend_items(input_items, rules)
print("Recommended items")
for item in output:
    print(str(item))
```
